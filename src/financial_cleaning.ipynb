```python
import numpy as np
import pandas as pd
import re
```

```python
finance = pd.read_csv(r"E:\.A Data Analyst\Excel\financial data\Loan_Default.csv\finance.csv")
finance.columns.tolist()
```

```python
finance = finance.rename(columns={'ID': 'Customer_id','year':'Year','loan_limit':'Loan_limit'})
finance
```

```python
finance['Loan_limit']
```

```python
finance['Year'] = pd.date_range(
    start='2020-01-01', 
    end='2024-12-31', 
    periods=len(finance)
)
finance
```

```python
finance['Year'].info()
```

```python
finance['Loan_limit']=finance['Loan_limit'].replace({'cf': 'Credit Facility', 'ncf': 'Non-Credit Facility'})
```

```python
finance.rename(columns={'gender': 'Gender','approv_in_adv': 'AIA_Type','loan_type': 'Loan_type'}, inplace=True)
finance['Gender'] = finance['Gender'].replace({'Sex Not Available': 'Unknown'})
finance['AIA_Type'] = finance['AIA_Type'].replace({'pre': 'Prep', 'nopre': 'Non-Prep'})
```

```python
finance
```

```python
finance['Loan_type']=finance['Loan_type'].replace({ 'type1':'Personal Loan','type2':'Business Loan','type3':'Home Loan'})
```

```python
finance=finance.rename(columns={'loan_purpose':'Loan_purpose'})
```

```python
def loan_purpose_function(row):
    if row['Loan_type']== 'Personal Loan':
        if row['Loan_purpose']== 'p1':
            return 'Education'
        else:
            return 'Medical Needs'
    elif row['Loan_type'] == 'Business Loan':
        return 'Business Expansion'
    elif row['Loan_type'] == 'Business Loan':
        return 'Business Expansion'
    elif row['Loan_type'] == 'Home Loan':
        return 'Housing'
    return row['Loan_purpose']
    

```

```python
finance['Loan_purpose']=finance.apply(loan_purpose_function,axis=1)
```

```python
finance['Credit_Worthiness']=finance['Credit_Worthiness'].replace({'l1':'Low Risk','l2':'High Risk'})
```

```python
finance['open_credit']=finance['open_credit'].replace({'nopc':'No Open Credit', 'opc':'Open Credit'})
```

```python
finance['business_or_commercial']=finance['business_or_commercial'].replace({'b/c':'business or commercial','nob/c':' non-business or commercial'})
```

```python
finance.columns.tolist()
```

```python
finance=finance.rename(columns={'rate_of_interest':'intrest_rate'})
```

```python
finance['intrest_rate']=finance['intrest_rate'].fillna(finance['intrest_rate'].median())
```

```python
finance['Interest_rate_spread']=finance['Interest_rate_spread'].fillna(finance['Interest_rate_spread'].median())
```

```python
finance['Interest_rate_spread'].isnull().sum()
```

```python
finance['Upfront_charges']=finance['Upfront_charges'].fillna(0.0)
```

```python
finance['property_value']=finance['property_value'].fillna(finance['loan_amount'].median())
```

```python
finance['income']=finance['income'].fillna(finance['income'].median())
```

```python
finance['LTV']=finance['LTV'].fillna(finance['LTV'].median())
```

```python
finance=finance.rename(columns={'dtir1':'dept_to_income_ratio'})
finance.tail()
```

```python
finance['dept_to_income_ratio']=finance['dept_to_income_ratio'].fillna(finance['dept_to_income_ratio'].median())
```

```python
finance['Loan_Tenure']=finance['term']/12
```

```python
finance['Loan_Tenure']=finance['Loan_Tenure'].fillna('NaN')
```

```python
finance['Loan_Tenure'].isnull().sum()
```

```python
finance['Neg_ammortization']=finance['Neg_ammortization'].fillna('NaN')
```

```python
finance['construction_type']=finance['construction_type'].replace({'sb':'Site Built','mh':'Manufactured Home'})
```

```python
finance['interest_only']=finance['interest_only'].replace({'not_int':'Not Intrest Only','int_only':'Intrest Only'})
```

```python
finance['lump_sum_payment']=finance['lump_sum_payment'].replace({'lpsm':'Lumpsum','not_lpsm':'Not Lumpsum'})
```

```python
finance['occupancy_type']=finance['occupancy_type'].replace({'pr':'Primary Residence', 'sr': 'Secoundary Residence', 'ir':'Investmnet Residence'})
```

```python
finance['credit_type']=finance['credit_type'].replace({'EXP':'Experian', 'EQUI':'Equifax', 'CRIF':'CRIF High Mark', 'CIB':'TransUnion CIBIL'})
```

```python
finance['co-applicant_credit_type']=finance['co-applicant_credit_type'].replace({'CIB':'TransUnion CIBIL', 'EXP':'Experian'})
```

```python
finance['age'].info()
```

```python
finance['submission_of_application']=finance['submission_of_application'].replace({'to_inst':'To be instructed','not_inst':'Not instructed', 'nan':'NaN'})
```

```python
finance['submission_of_application']=finance['submission_of_application'].fillna('NaN')
```

```python
finance['submission_of_application'].isnull().sum()
```

```python
finance['Region']=finance['Region'].replace({'south':'South','central':'Central'})
```

```python
finance['Security_Type']=finance['Security_Type'].replace({'direct':'Direct'})
```

```python
finance['Status']= finance['Status'].map({1: 'Active', 0: 'Inactive'})
```

```python
def credit_score_band (Credit_Score):
    if Credit_Score>=750:
        return "Excellent"
    elif Credit_Score>=700:
        return "Good"
    elif Credit_Score>=650:
        return "Average"
    else:
        return "Poor"
```

```python
finance['Credit_Score_Band'] = finance['Credit_Score'].apply(credit_score_band)
```

```python
def risk_band(LTV):
    if LTV<= 60:
        return "Low Risk"
    elif LTV<= 80:
        return "Moderate Risk"
    elif LTV<=90:
        return "High Risk"
    else:
        return "Very High Risk"
```

```python
finance['Risk_Bucket']=finance['LTV'].apply(risk_band)
```

```python
finance['Exposure_amount']=finance['loan_amount']
```

```python
finance['Loan_Tenure'] = finance['Loan_Tenure'].replace('NaN', np.nan)
```

```python
finance['Loan_Tenure'] = finance['Loan_Tenure'].fillna(finance['Loan_Tenure'].median())
```

```python
finance['Interest_Income_Estimated']=finance['Exposure_amount']*finance['intrest_rate']*finance['Loan_Tenure']
```

```python
finance['Interest_Income_Estimated']
```

```python
def ltv_flag(LTV):
    if LTV > 80:
        return 1
    else:
        return 0
```

```python
finance['high_ltv_flag'] = finance['LTV'].apply(ltv_flag)
```

```python
def dtir_flag(dept_to_income_ratio):
    if dept_to_income_ratio > 40:
        return 1
    else:
        return 0
```

```python
finance['DITR_Flag']=finance['dept_to_income_ratio'].apply(dtir_flag)
```

```python
finance['DITR_Flag']
```

```python
r = finance['intrest_rate'] / 12 #Monthly_intrest_rate
n= finance['Loan_Tenure'] * 12 #Total_number_of_monthly_installment

finance['monthly_installment'] = finance['loan_amount']*(r*(1+r)**n)/((1 + r) ** n - 1)
```

```python
finance['Loan_limit']=finance['Loan_limit'].fillna('NaN')
```

```python
finance['AIA_Type']=finance['AIA_Type'].fillna('NaN')
finance['age']=finance['age'].fillna('NaN')
```

```python
finance['term']=finance['term'].fillna(finance['term'].median())
finance['monthly_installment']=finance['monthly_installment'].fillna(finance['monthly_installment'].median())
```
